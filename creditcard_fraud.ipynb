## Build model

```python
X = df.drop(['id', 'Class'], axis=1)
y = df['Class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

## Model 1: Logistic Regression

```python
m1 = linear_model.LogisticRegression()
m1.fit(X_train, y_train)
#Predict the response for test dataset
y_pred_train_m1 = m1.predict(X_train)
y_pred_test_m1 = m1.predict(X_test)
```

```python
print(classification_report(y_train,y_pred_train_m1,target_names=["no Fraud", "Fraud"]))
# print(f'accuracy score =',accuracy_score(y_train,m1.predict(X_train)))
```

```python
print(classification_report(y_test,y_pred_test_m1,target_names=["no Fraud", "Fraud"]))
# print(f'accuracy score =',accuracy_score(y_test,m1.predict(X_test)))
```

## Model 2: Dicision Tree

```python
m2 =  DecisionTreeClassifier(criterion= "gini", max_depth=None)

#Step 3: Train Decision Tree Classifer
m2 = m2.fit(X_train,y_train)
#Predict the response for test dataset
y_pred_train_m2 = m2.predict(X_train)
y_pred_test_m2 = m2.predict(X_test)
```

```python
print(classification_report(y_train,y_pred_train_m2,target_names=["no Fraud", "Fraud"]))
print(f"accuracy score = {accuracy_score(y_train,y_pred_train_m2)}")
```

```python
print(classification_report(y_test,y_pred_test_m2,target_names=["no Fraud", "Fraud"]))
print(f'accuracy score =',accuracy_score(y_test,y_pred_test_m2))
```
